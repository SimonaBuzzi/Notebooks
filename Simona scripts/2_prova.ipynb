```python
 ## preprocessing is a module within the sklearn library. The preprocessing module in scikit-learn provides various functions and classes for data preprocessing tasks in machine learning. StandardScaler is a class within the preprocessing module. Specifically, it's a class used for standardizing features by removing the mean and scaling to unit variance###
```

```python
# Weight and Volume are scales which means z = x-u/s 
# fit looks for u and s based on the features of the data
# transform transforms the data 
# scaledX has the two methods combined
import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler() # the class behaves like a function(?)

df = pandas.read_csv("Downloads\data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

scaledX = scale.fit_transform(X) #scale.fit(x) and scale.transform(x) theyre combined

regr = linear_model.LinearRegression()
regr.fit(scaledX, y)

scaled = scale.transform([[2300, 1.3]]) #scale.fit(x) and scale.transform(x) 
                                        # this is z = x-u/s # mean and s are in the learner already

predictedCO2 = regr.predict([scaled[0]])
print(predictedCO2)
```

```python
import pandas as pd

df = pd.read_csv("Downloads/Book1.csv")

print(df)
```

```python
d = {'UK': 0, 'USA': 1, 'N': 2} #d= dictionary, UK, USa etc. are "Kyes"
df['Nationality'] = df['Nationality'].map(d) # map()replaces YES with 1 and NO with 0
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

print(df)
```

```python
# Then we have to separate the feature columns from the target column.

#The feature columns are the columns that we try to predict from, and the target column is the column with the values we try to predict.
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality'] #from where we build 

X = df[features]
y = df['Go'] # target column, where we want to do the predictions

print(X)
```

```python
print(y)
```

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

dtree = DecisionTreeClassifier() # create the instance. dtree is an arbitrary name
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

```python
print(dtree.predict([[40, 10, 7, 1]]))
```

```python
import matplotlib.pyplot as plt
import numpy
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
numpy.random.seed(2)
actual = numpy.random.binomial(1,.9,size = 1000)
predicted = numpy.random.binomial(1,.9,size = 1000)

# Compute the confusion matrix
conf_matrix = confusion_matrix(actual, predicted)

# Create an instance of ConfusionMatrixDisplay
cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=[False, True])
cm_display.plot()
plt.show()
```
