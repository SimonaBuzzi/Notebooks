```python
import numpy as np
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = np.mean(speed)
y = np.median(speed)

x_int = int(x)
y_int = int(y)

print(x_int)
print(y_int)
```

```python
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86] #count array, how many times the value appears in the array
x = stats.mode(speed, keepdims=True)
print(x)
```
